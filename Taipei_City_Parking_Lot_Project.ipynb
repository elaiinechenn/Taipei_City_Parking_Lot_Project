# 未波動累積分布圖

```
df_normal
```

```
df_normal['updatetime'] = pd.to_datetime(df_normal['updatetime'],utc = True)
df_normal['updatetime'] = pd.to_datetime(df_normal['updatetime']).apply(lambda x: x.replace(tzinfo=None))
```

```
#將時間跟id分類後才可以做shift不然不同月份會拿到上一筆的資料
df_normal = df_normal.sort_values(['id', 'updatetime'])
df_normal['shift'] = df_normal.groupby(['id'])['availablecar'].shift()
df_normal.iloc[8485:8500] #檢查
```

```
#df['data']= df['updatetime'].dt.date 取同一天 date
# df_normal['week']= df_normal['updatetime'].dt.dayofweek #可以計算是禮拜幾
```

```
#刪除12/31的資料
df_normal.drop(df_normal.loc[df_normal['updatetime'] == '2020-12-31 23:56:00'].index, inplace=True) 
```

```
# 未波動是前一筆減上一筆為0的值,加絕對值避免其實有波動但抵銷的情況
df_normal['variation'] = abs(df_normal['availablecar']-df_normal['shift'])
df_normal['data']= df_normal['updatetime'].dt.date
df_normal.iloc[8485:8500]
```

```
#依據一天將同個id的變化值加總 0就等於一天都沒有波動
df_normal_num = df_normal.groupby(['data','id']).variation.sum().reset_index()
df_normal_num
```

```
#標記variation=0的為1
df_normal_num['innor_data'] = 1
df_normal_num.loc[df_normal_num['variation'] != 0 ,'innor_data'] = 0
df_normal_num.sort_values('id')
```

```
total_df_normal = len(df_normal_num['data'])
total_innor_data = df_normal_num['innor_data'].sum()
print('全部資料數:',total_df_normal,'異常(未波動)的資料:',total_innor_data)
```

```
#查看同個停車場整天沒有變化的天數,59代表兩個月都沒有動
finally_num = df_normal_num.groupby('id').innor_data.sum().reset_index()
finally_num
```

```
y_l = [i/100 for i in range(101)]
x_l = finally_num['innor_data'].quantile(y_l)

p = figure(title='未波動累積分布圖', plot_width=600, plot_height=600)
p.circle(x_l, y_l, color="purple")
p.line(x_l, y_l, color="pink")
p.xaxis.axis_label = '停車場未波動的天數'
p.yaxis.axis_label = '停車場百分比'   
show(p)
```

# 劇烈波動累積分布圖

```
df_max = df_clear.groupby('id').availablecar.max().reset_index()
df_normal_max = pd.merge(df_clear,df_max,on = ['id'])
df_normal_max
```

```
df_normal_max = df_normal_max.rename(columns={'availablecar_y': 'max_park','availablecar_x': 'availablecar'})
```

```
#劇烈波動的依據是停車位變化量除最大停車位數 = high_var
df_normal_max['high_var']= abs(df_normal_max['variation']) / df_normal_max['max_park']
df_normal_max.sort_values('high_var')
```

```
#處理nan值
df_normal_max = df_normal_max.fillna(0)
df_normal_max.sort_values('high_var')
```
